```python
alunos = {}
# Dicionário que armazena todos os alunos
def cadastrar_alunos(alunos):
    try:
        print("\n-------- CADASTRO DE ALUNO -------")
        nome = input("Informe o nome do aluno: ")
        if not nome.replace(" ", "").isalpha():
            raise ValueError("Nome deve conter apenas letras e espaços")

        idade = int(input("Informe sua idade: "))
        if idade < 0 or idade > 120:
            raise ValueError("Idade deve estar entre 0 e 120")

        curso = input("Informe seu curso (ADS, Eng Computação, Ciência Computação): ").title()
        if curso not in["Ads", "Eng Computação", "Ciência Computação"]:
            raise ValueError("Curso dever ser ADS, Eng Computação ou Ciência Computação")

        notafinal = float(input("Informe sua Nota Final: "))
        if notafinal < 0 or notafinal > 10:
            raise ValueError("Nota deve estar entre 0 e 10")

        # dicionario que armazena cada aluno

        aluno = {
            "Nome":nome,
            "Idade":idade,
            "Curso":curso,
            "Nota Final":notafinal
        }

        matricula = len(alunos) + 1

        alunos[matricula] = aluno

        print("Aluno Cadastrado!!")
    except ValueError as e:
        print(f"Erro no cadastro: {e}")
    except Exception  as e:
        print(f"Ocorreu um erro inesperado: {e}")
    finally:
        print("Operaçaõ de cadastro concluída")

def listar_alunos(alunos, *campos):

    if not campos: # caso nenhum campo seja mostrado, mostre ...
        campos = ("Nome", "Idade", "Curso", "Nota Final")

    for matricula, aluno in alunos.items():
        print(f"\nMatrícula: {matricula}")
        for campo in campos:
            print(f"{campo}: {aluno.get(campo, 'Campo inválido')}")

def listar_aprovados(alunos, **criterio): #  , ads = 9
    try:
        if not alunos:
            print("Nenhum aluno cadastrado!")
            return

        notas_corte = {
            "Ads": 7,
            "Eng Computação": 7,
            "Ciência Computação": 7
        }

        # Atualiza as notas de corte se especificado nos criterios
        for curso, nota in criterio.items(): # para cada curso e nota em Ads = 9, logo (curso = Ads e nota = 9)
            curso_formatado = curso.title() # curso_form. = Ads
            if curso_formatado in notas_corte: # se Ads está dentro do dicionario notas_corte
                notas_corte[curso_formatado] = nota # notas_corte[Ads] = nota (9), logo Ads = 9 e modifiquei a nota de corte desse curso


        print("\nAlunos Aprovados: ")
        for matricula, aluno in alunos.items():
            curso = aluno["Curso"]
            notafinal = aluno["Nota Final"]
            notas_corte = notas_corte.get(curso, 7)

            if notafinal >= notas_corte:
                print(f"\nMatrícula {matricula}")
                print(f"Nome {aluno['Nome']}")
                print(f"Curso: {curso}")
                print(f"Nota Final: {notafinal} (Corte: {notas_corte})")
    except Exception as e:
        print(f"Erro ao listar aprovados:{e}")

while True:
    try:
        print("\nMENU PRINCIPAL")
        print("1. Cadastrar aluno")
        print("2. Listar todos os alunos")
        print("3. Listar alunos aprovados")
        print("4. Sair ")

        op = input("Qual Operação irá realizar:")


        if op == "1":
            cadastrar_alunos(alunos)  # ← Chamando função com nome diferente da definição
        elif op == "2":
    # Pode passar campos especificos, listar_alunos(alunos, "Nome", "Nota Final")
            listar_alunos(alunos)
        elif op == "3":
    # Pode passar campos especificos, listar_alunos(alunos, Ads = 9)
            listar_aprovados(alunos, ads=9)
        elif op == "4":
            print("Saindo do Sistema!")
            break
        else:
            print("Operação inválida")
    except Exception as e:
        print(f"Ocorreu um erro {e}")
















```
